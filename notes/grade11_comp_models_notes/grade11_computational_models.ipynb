# مقدّمة

* موضوع النماذج الحسابية هو موضوع نظري، لا يوجد فيه برمجة.
* هدف الموضوع هو وصف الحاسوب بطريقة رياضية حتى نستطيع فحص حدود إمكانياته.
* هل يستطيع الحاسوب حل أي مشكلة حسابية؟ (الجواب لا طبعا)
* قبل أن نصل إلى نموذج الحاسوب (ماكنة تيورنج) نتعرف على نماذج أضعف من الحاسوب (قدرتها الحسابية أقل من قدرة الحاسوب)
* الموديل هو تخطيط يصف طريقة عمل شيء معين.
* مهمة حسابية هي مهمة هدفها معالجة المدخلات حتى تعطي نتيجة بناء على هذه المدخلات.
* في موضوع نماذج حسابية، نهتم بالمشاكل الحسابية التي تكون نتيجتها ثنائية (نعم أو لا - مقبول أو مرفوض...)

**مثال 2**: موقف سيارات له بوابتان، يمكن الدخول فقط إذا كانتا مفتوحتين في نفس الوقت.

<img src="../../images/DFAs/02_two_gates_to_open_dfa.png" alt="lamp automata" width="500" height="250" style="display:block; margin:auto;">

* `open 1` يعني فتح البوابة الأولى
* `close 1` يعني إغلاق البوابة الأولى
* `open 2` يعني فتح البوابة الثانية
* `close 2` يعني إغلاق البوابة الثانية

* $q_0$ يمثل الحالة التي فيها كلا البوابتين مغلقتان
* $q_1$ يمثل الحالة التي فيها البوابة 1 مفتوحة، والبوابة 2 مغلقة
* $q_1$ يمثل الحالة التي فيها البوابة 1 مغلقة، والبوابة 2 مفتوحة
* $q_3$ يمثل الحالة التي فيها كلا البوابتين مفتوحتان

<div style="page-break-after: always;"></div>

**مثال 3**: هناك حاجز مكون من خشبات واقفة بشكل عمودي بجانب بعضها، كل خشبة مدهونة بلون واحد من الألوان أخضر، أحمر، أزرق. نريد بناء أوتومات يستقبل ألوان الخشبات للحاجز، ويقرر هل حالة دهن هذا الحاجز مقبول أم لا؟
نعتبر الحاجز مقبولا فقط إذا لم تكن هناك خشبتان متجاورتان لهما نفس اللون.

نبني موديلا يقرأ ألوان الخشبات بالترتيب. ونرمز للألوان بالأحرف: R للأحمر، G للأخضر، B للأزرق.

<img src="../../images/DFAs/03_color_series_dfa.png" alt="lamp automata" width="500" height="250" style="display:block; margin:auto;">

* الذي يهمنا في هذه الحالة هو: ما هو آخر لون رأيناه؟ الألوان التي قبل ذلك لا تهمّنا.
* كل واحد من الحالات $q_1, q_2, q_3$ يمثل اللون الأخير الذي رأيناه:
    - $q_1$ يشير إلى أنّ اللون الأخير الذي رأيناه هو الأحمر
    - $q_2$ يشير إلى أنّ اللون الأخير الذي رأيناه هو الأخضر
    - $q_3$ يشير إلى أنّ اللون الأخير الذي رأيناه هو الأزرق
* في كل واحد من هذه الحالات الثلاثة: إذا رأينا اللون الذي تمثله هذه الحالة مرة أخرى، فإنّنا نكون رأينا نفس اللون مرتين متتاليتين، وبالتالي فإنّ هذا المُدخَل غير قانوني (يوجد فيه خشبتان متجاورتان لهما نفس اللون).
في هذه الحالة ننتقل إلى $q_4$ الذي يمثل حالة الفخّ، أي أنّنا لن نخرج منها أبدًا، وسنحكم على المدخل بأنّه غير قانوني.
* بالمقابل، إذا لم نصل إلى الحالة $q_4$ فإنّ هذا يعني أنّ المدخل كان قانونيًا ومقبولًا - لذلك نجعل الحالات الثلاثة $q_1, q_2, q_3$ كلها حالات قبول (ويظهر هذا في الرسم بجعلها دوائر مزدوجة.)

## وصف للأوتومات

الأوتومات الأخير (مثال 3) يعتبر موديلا حسابيا، يحسب نتيجة مدخل معين وجوابه هو "نعم" (الدهن قانوني) أو "لا" (الدهن غير قانوني)

### تعريف الأوتومات رياضيًّا

نعرّف مثلًا الأوتومات في المثال 3:

يمكننا وصف الأوتومات بأنّه مكون من المجموعات التالية:

1. مجموعة الحالات:
$$Q = \{q0, q1, q2, q3, q4\}$$

2. الحالة الابتدائية:
$$Start = q0$$

3. مجموعة الحروف التي يتكون منها المدخل:
$$ \Sigma = \{R,G,B\}$$

4. الحالات القابلة
$$ F = \{q1, q2, q3\}$$

5. دالة الانتقالات بين الحالات:
$$\delta: \{Q \times \Sigma \} \to Q$$

هذه الدالة تستقبل: الحالة الحالية مع الحرف التالي في المدخل، وتخبرنا إلى أي حالة علينا الانتقال.
