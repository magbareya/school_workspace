# مقدّمة

* موضوع النماذج الحسابية هو موضوع نظري، لا يوجد فيه برمجة.
* هدف الموضوع هو وصف الحاسوب بطريقة رياضية حتى نستطيع فحص حدود إمكانياته.
* هل يستطيع الحاسوب حل أي مشكلة حسابية؟ (الجواب لا طبعا)
* قبل أن نصل إلى نموذج الحاسوب (ماكنة تيورنج) نتعرف على نماذج أضعف من الحاسوب (قدرتها الحسابية أقل من قدرة الحاسوب)
* الموديل هو تخطيط يصف طريقة عمل شيء معين.
* مهمة حسابية هي مهمة هدفها معالجة المدخلات حتى تعطي نتيجة بناء على هذه المدخلات.
* في موضوع نماذج حسابية، نهتم بالمشاكل الحسابية التي تكون نتيجتها ثنائية (نعم أو لا - مقبول أو مرفوض...)

### لغة الأوتومات

عندما نشغّل الأوتومات على مُدخَل معيّن، فإنّنا نقول إن الأوتومات **قبل المدخل** أو **قبل الكلمة** إذا كان الأوتومات موجودًا في حالة قابلة عند الانتهاء من قراءة الكلمة (المدخل).

**لغة الأوتومات** هي مجموعة كل الكلمات التي يقبلها الأوتومات.

مثلًا، في المثال 3 أعلاه: لغة الأوتومات هي مجموعة كل الكلمات المكونة من الحروف R,G,B بحيث لا يوجد حرفان متشابهان متجاوران.

عادة، نرمز للغة الأوتومات بالحرف $L$.

# اللغات

## مقدمة في علم المجموعات (Set theory)

المجموعة هي تجميع لعناصر  (عادة تجمعها علاقة، أي تكون من نفس العالَم) نريد التعامل معها كوحدة واحدة. على سبيل المثال:
* مجموعة الأعداد الزوجية: {0, 2, 4, 6, …}
* مجموعة الأحرف العربية: {أ، ب، ت، ... ي}
* مجموعة الأرقام الصحيحة: {0، 1، 1-، 2، 2-، ...}

مثلًا، نعرّف المجموعات التالية
$$ S1 = \{ 1, 2, 3, 4, 5, 6 \} \\
   S2 = \{ 2, 4, 6 \} \\
   S3 = \{ 5, 6, 7 \} \\
   S4 = \{ 5, 6, 7 \} $$

### تعريفات ورموز في علم المجموعات
* المجموعة الخالية هي المجموعة التي لا تحتوي على أي عنصر، نرمز لها عادة بالحرف اليوناني $\Phi = \{\}$ (في).

* $a \in A$: العنصر ينتمي للمجموعة<br>
أي أنّ العنصر $a$ موجود في المجموعة A.<br>
مثلًا، في المجموعات أعلاه: $3 \in S1$

* $a \notin A$: العنصر **لا** ينتمي للمجموعة <br>
أي أنّ العنصر $a$ **غير موجود** في المجموعة A. <br>
مثلًا، في المجموعات أعلاه: $7 \notin S1$

* $A \subseteq B$: مجموعة تحتوي مجموعة أخرى (أو: مجموعة هي مجموعة جزئية لمجموعة أخرى) <br>
أي أن المجموعة $B$ تحتوي المجموعة $A$. بكلمات أخرى: كل عنصر موجود في المجموعة $A$ فهو بالضرورة موجود أيضًا في المجموعة $B$.<br>
مثلًا، في المجموعات أعلاه: $S2 \subseteq S1$

* $A \nsubseteq B$: مجموعة **لا** تحتوي مجموعة أخرى (أو: مجموعة هي ليست مجموعة جزئية لمجموعة أخرى) <br>
أي أنّ هناك عناصر موجودة في $A$ لكنها غير موجودة في $B$. <br>
مثلًا، في المجموعات أعلاه: $S3 \nsubseteq S1$

* $A = B$: مساواة بين مجموعتين <br>
أي أنّ كل العناصر الموجودة في $A$ موجودة في $B$، والعكس أيضًا: كل عنصر في $B$ موجود في $A$ أيضًا. <br>
مثلًا، في المجموعات أعلاه: $S3 = S4$

* $|A|$: حجم المجموعة <br>
أي عدد العناصر المختلفة في هذه المجموعة. <br>
مثلًا، في المجموعات أعلاه: $|S1| = 6$ و $|S2| = |S3| = |S4| = 3$.

* $A \cup B$: اتحاد مجموعتين <br>
أي المجموعة المكونة من كل العناصر الموجودة في $A$ مع كل العناصر الموجودة في $B$. <br>
مثلًا، في المجموعات أعلاه: $S2 \cup S3 = \{2, 4, 5, 6, 7 \}$.

* $A \cap B$: تقاطع مجموعتين <br>
أي المجموعة المكونة من كل العناصر الموجودة في المجموعة $A$ وفي المجموعة $B$ في نفس الوقت. <br>
مثلًا، في المجموعات أعلاه: $S2 \cap S3 = \{6 \}$. لأنّه فقط العنصر 6 موجود في كلا المجموعتين.

* $A \setminus B$: الفرق بين مجموعتين <br>
أي المجموعة المكونة من كل العناصر الموجودة في المجموعة $A$ وليست موجودة في المجموعة $B$. <br>
مثلًا، في المجموعات أعلاه: $S2 \setminus  S3 = \{2, 4 \}$. لأنّه فقط العنصر 6 موجود في كلا المجموعتين.

**:ملاحظات**
- المجموعة الخالية هي مجموعة جزئية لكل مجموعة أخرى $\Phi \subseteq A$ لكل مجموعة $A$.
- كل مجموعة هي مجموعة جزئية لنفسها $A \subseteq A$.
- في المجموعات لا توجد أهمية للترتيب
- في المجموعات كل عنصر يظهر مرة واحدة (لا يوجد تكرار)

## رموز وتعريفات في اللغات

### الأبجدية
الأبجدية هي مجموعة نهائية وغير فارغة من الرموز (الرموز يمكن أن تكون أي رموز: أرقام، أحرف، علامات ترقيم...).

عادة نرمز للأبجدية بالحرف $\Sigma$ (سيجما).

مثلًا: 
* $\Sigma = \{a, b\}$ هي أبجدية مكونة من رمزين اثنين (حرفَين)
* $\Sigma = \{0, 1\}$ هي أبجدية مكونة من رمزين اثنين أيضًا (في هذه الحالة رقمين)
* $\Sigma = \{0, 1, 2, \dots, 9\}$ هي أبجدية مكونة من عشرة رموز (الأرقام).


<br>

### الكلمة

الكلمة هي تسلسل من الرموز من أبجدية معينة.

* مثلًا، بالنسبة للأبجدية $\Sigma = \{0, 1\}$ فإن السلاسل التالية هي كلمات مكونة من هذه الأبجدية:
    * $w_1 = 000$
    * $w_2 = 111000$
    * $w_3 = 1010$
    * $w_4 = 01011$

* نستحدم الرمز $|w|$ للتعبير عن عدد الرموز في الكلمة $w$.
مثلًا، في الكلمات أعلاه:
    * $|w_1| = 3$
    * $|w_2| = 6$
    * $|w_3| = 4$
    * $|w_4| = 5$

* الكلمة قد تكون فارغة، أي لا يوجد فيها أي رموز. نرمز للكلمة فارغة بالحرف $\epsilon$ (إبسيلون).  <br>
حسب التعريف في النقطة السابقة: $|\epsilon| = 0$ لأنّ عدد الرموز في الكلمة الفارغة هو 0.


### اللغة
اللغة هي **مجموعة من الكلمات** المعرّفة فوق أبجدية معينة.

أمثلة:
- $L_1 = \{0, 1, 01, 10\}$ معرفة فوق الأبجدية $\Sigma = \{0, 1\}$

اللغة يمكن أن تكون:
- **منتهية** (عدد كلماتها محدود).
- **غير منتهية** (عدد كلماتها غير محدود).

## عمليات على الأبجدية واللغات والكلمات

### عملية القوة للأبجدية
نعرف عمليتين أساسيتين:

### 1. $\Sigma^n$
جميع الكلمات الممكنة من الأبجدية $\Sigma$ بطول $n$.

مثال: إذا كانت $\Sigma = \{0,1\}$:
- $\Sigma^1 = \{0, 1\}$
- $\Sigma^2 = \{00, 01, 10, 11\}$
- $\Sigma^3 = \{000, 001, 010, 011, 100, 101, 110, 111\}$

### 2. $\Sigma^*$
جميع الكلمات الممكنة من أي طول (بما في ذلك الكلمة الفارغة) فوق الأبجدية $\Sigma$.

مثال: $\Sigma = \{a, b\}$
$\Sigma^* = \{\epsilon, a, b, aa, ab, ba, bb, aaa, aab, \dots\}$

<br>

### ضرب / إلصاق الكلمات واللغات
- **ضرب كلمتين:** إذا كان $u = 01$ و $v = 10$، فإن $uv = 0110$.
- **ضرب لغتين:** إذا كان $L_1 = \{0,1\}$ و $L_2 = \{a,b\}$:
$L_1L_2 = \{0a, 0b, 1a, 1b\}$

### القوة للغة
- $L^n$ تعني ضرب اللغة بنفسها $n$ مرات.

مثال: إذا كانت $L = \{a,b\}$:
- $L^1 = \{a,b\}$
- $L^2 = \{aa, ab, ba, bb\}$
- $L^3 = \{aaa, aab, aba, abb, baa, bab, bba, bbb\}$
