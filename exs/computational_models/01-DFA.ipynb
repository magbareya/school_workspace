### القسم الثاني: بناء أوتومات من اللغة

لكل لغة من اللغات التالية، ابنِ أوتوماتًا يقبل هذه اللغة.

1.  **اللغة:** كل السلاسل من الأبجدية $\{0, 1\}$ التي لها عدد زوجي من الأصفار.

2.  **اللغة:** كل السلاسل من الأبجدية $\{a, b\}$ التي تبدأ وتنتهي بنفس الحرف.

3.  **اللغة:** كل السلاسل من الأبجدية $\{0, 1\}$ التي طولها يقبل القسمة على 3.

4.  **اللغة:** كل السلاسل من الأبجدية $\{a, b, c\}$ التي لا تحتوي على السلسلة الفرعية 'ac'.

5.  **اللغة:** كل السلاسل من الأبجدية $\{a, b\}$ التي يكون الحرف الثالث من النهاية هو 'a'.

6.  **اللغة:** كل السلاسل من الأبجدية $\{0, 1\}$ التي تحتوي على '1' واحدة على الأقل.

7.  **اللغة:** كل السلاسل من الأبجدية $\{a, b\}$ التي لها عدد فردي من الحرف 'a' وعدد فردي من الحرف 'b'.

8.  **اللغة:** اللغة التي تحتوي على السلسلة "baba" فقط.

9.  **اللغة:** كل السلاسل من الأبجدية $\{0, 1\}$ حيث كل '0' تتبعها مباشرة '1'.

10. **اللغة:** كل السلاسل من الأبجدية $\{a, b\}$ التي لا تنتهي بـ 'ab'.
