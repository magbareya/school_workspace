# تمارين على الأوتومات النهائي المحدد

### القسم الأول: تحديد اللغة من DFA

لكل من ذاتيات الحالة المنتهية المحددة (DFA) التالية، حدد اللغة التي تقبلها.

1.  **DFA تبدأ بحرف معين:**
    هذه الـDFA تقبل السلاسل التي تبدأ بالحرف 'a'.
    * الحالات: $q_0$ (بداية)، $q_1$ (قبول)، $q_{reject}$
    * الانتقالات: $q_0 \xrightarrow{a} q_1$؛ $q_0 \xrightarrow{b} q_{reject}$؛ $q_1 \xrightarrow{a, b} q_1$؛ $q_{reject} \xrightarrow{a, b} q_{reject}$

2.  **DFA تنتهي بحرف معين:**
    هذه الـDFA تقبل السلاسل التي تنتهي بالحرف 'b'.
    * الحالات: $q_0$ (بداية)، $q_1$ (قبول)
    * الانتقالات: $q_0 \xrightarrow{a} q_0$؛ $q_0 \xrightarrow{b} q_1$؛ $q_1 \xrightarrow{a} q_0$؛ $q_1 \xrightarrow{b} q_1$

3.  **DFA تحتوي على سلسلة فرعية معينة:**
    هذه الـDFA تقبل السلاسل التي تحتوي على السلسلة الفرعية 'ab'.
    * الحالات: $q_0$ (بداية)، $q_1$, $q_2$ (قبول)
    * الانتقالات: $q_0 \xrightarrow{a} q_1$؛ $q_0 \xrightarrow{b} q_0$؛ $q_1 \xrightarrow{a} q_1$؛ $q_1 \xrightarrow{b} q_2$؛ $q_2 \xrightarrow{a, b} q_2$

4.  **DFA لطول محدد:**
    هذه الـDFA تقبل السلاسل التي طولها بالضبط 2.
    * الحالات: $q_0$ (بداية)، $q_1$, $q_2$ (قبول)
    * الانتقالات: $q_0 \xrightarrow{a, b} q_1$؛ $q_1 \xrightarrow{a, b} q_2$

5.  **DFA لعدد زوجي من الحروف:**
    هذه الـDFA تقبل السلاسل التي لها عدد زوجي من الحروف.
    * الحالات: $q_0$ (بداية، قبول)، $q_1$
    * الانتقالات: $q_0 \xrightarrow{a, b} q_1$؛ $q_1 \xrightarrow{a, b} q_0$

6.  **DFA لعدد فردي من الحروف:**
    هذه الـDFA تقبل السلاسل التي لها عدد فردي من الحروف.
    * الحالات: $q_0$ (بداية)، $q_1$ (قبول)
    * الانتقالات: $q_0 \xrightarrow{a, b} q_1$؛ $q_1 \xrightarrow{a, b} q_0$

7.  **DFA لا تحتوي على حرف معين:**
    هذه الـDFA تقبل السلاسل التي لا تحتوي على الحرف 'b'.
    * الحالات: $q_0$ (بداية، قبول)، $q_{reject}$
    * الانتقالات: $q_0 \xrightarrow{a} q_0$؛ $q_0 \xrightarrow{b} q_{reject}$؛ $q_{reject} \xrightarrow{a, b} q_{reject}$

8.  **DFA للغة الفارغة:**
    هذه الـDFA لا تقبل أي سلسلة.
    * الحالات: $q_0$ (بداية)، $q_{reject}$
    * الانتقالات: $q_0 \xrightarrow{a, b} q_{reject}$؛ $q_{reject} \xrightarrow{a, b} q_{reject}$

9.  **DFA تحتوي على 1 على الأقل:**
    هذه الـDFA تقبل السلاسل من الأبجدية $\{0, 1\}$ التي تحتوي على الرقم '1' على الأقل.
    * الحالات: $q_0$ (بداية)، $q_1$ (قبول)
    * الانتقالات: $q_0 \xrightarrow{0} q_0$؛ $q_0 \xrightarrow{1} q_1$؛ $q_1 \xrightarrow{0, 1} q_1$

10. **DFA تقبل كل السلاسل:**
    هذه الـDFA تقبل أي سلسلة.
    * الحالات: $q_0$ (بداية، قبول)
    * الانتقالات: $q_0 \xrightarrow{a, b} q_0$

### القسم الثاني: بناء DFA من اللغة

لكل لغة من اللغات التالية، قم ببناء ذاتية حالة منتهية محددة (DFA) تقبل هذه اللغة.

1.  **اللغة:** كل السلاسل من الأبجدية $\{0, 1\}$ التي لها عدد زوجي من الأصفار.

2.  **اللغة:** كل السلاسل من الأبجدية $\{a, b\}$ التي تبدأ وتنتهي بنفس الحرف.

3.  **اللغة:** كل السلاسل من الأبجدية $\{0, 1\}$ التي طولها يقبل القسمة على 3.

4.  **اللغة:** كل السلاسل من الأبجدية $\{a, b, c\}$ التي لا تحتوي على السلسلة الفرعية 'ac'.

5.  **اللغة:** كل السلاسل من الأبجدية $\{a, b\}$ التي يكون الحرف الثالث من النهاية هو 'a'.

6.  **اللغة:** كل السلاسل من الأبجدية $\{0, 1\}$ التي تحتوي على '1' واحدة على الأقل.

7.  **اللغة:** كل السلاسل من الأبجدية $\{a, b\}$ التي لها عدد فردي من الحرف 'a' وعدد فردي من الحرف 'b'.

8.  **اللغة:** اللغة التي تحتوي على السلسلة "baba" فقط.

9.  **اللغة:** كل السلاسل من الأبجدية $\{0, 1\}$ حيث كل '0' تتبعها مباشرة '1'.

10. **اللغة:** كل السلاسل من الأبجدية $\{a, b\}$ التي لا تنتهي بـ 'ab'.
